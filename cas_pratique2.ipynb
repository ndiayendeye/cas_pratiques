## Question 2:

```python
Tableau=Notes[Notes.columns[2:]].values
Tableau
```

## physique

```python
NotesPhysique=Tableau[:,0:3]
Coef1=[2.5,2.5,1]
for j in range(len(NotesPhysique)):
    for i in range(0,3):
        NotesPhysique[j,i]=NotesPhysique[j,i]*Coef1[i]
NotesPhysique
moyen_Phy=np.sum(NotesPhysique,axis=1)/6

#Validation 
V1=[]
credit_Phy=[]
for i in range(len(moyen_Phy)):
    if moyen_Phy[i] >=10:
        V1.append("V")
        credit_Phy.append(12)
        

    else:
        V1.append("NV")
        credit_Phy.append(0)
V1
```

## MATHS

```python
NotesMaths=Tableau[:,3:5]

Coef2=[2.5,2.5]
for j in range(len(NotesMaths)):
    for i in range(2):
        NotesMaths[j,i]=NotesMaths[j,i]*Coef2[i]
moyen_Maths=np.sum(NotesMaths, axis=1)/5
V2=[]
credit_Math=[]
for i in range(len(moyen_Maths)):
    if moyen_Maths[i] >=10:
        V2.append("V")
        credit_Math.append(10)
    else:
        V2.append("NV")
        credit_Math.append(0)
V2
```

## Moyenne Chimie

```python
NotesChimie=Tableau[:,5:8]
Coef3=[2.5,2.5,1]
for j in range(len(NotesChimie)):
    for i in range(3):
        NotesChimie[j,i]=NotesChimie[j,i]*Coef3[i]
moyen_Chimie=np.sum(NotesChimie, axis=1)/6
V3=[]
credit_Chimie=[]
for i in range(len(moyen_Chimie)):
    if moyen_Chimie[i] >=10:
        V3.append("V")
        credit_Chimie.append(12)
    else:
        V3.append("NV")
        credit_Chimie.append(0)
V3
```

## Moyenne Infos

```python
NotesInfos=Tableau[:,8:10]
Coef4=[2.5,2.5]
for j in range(len(NotesInfos)):
    for i in range(2):
        NotesInfos[j,i]=NotesInfos[j,i]*Coef4[i]

moyenne_Infos=np.sum(NotesInfos, axis=1)/5
V4=[]
credit_Info=[]
for i in range(len(moyenne_Infos)):
    if moyenne_Infos[i] >=10:
        V4.append("V")
        credit_Info.append(10)
    else:
        V4.append("NV")
        credit_Info.append(0)
V4
```

## Moyenne Outils_Base

```python
notes_Ob=Tableau[:,10:13]
Coef5=[1.5,1.5,1]
for j in range(len(notes_Ob)):
    for i in range(3):
        notes_Ob[j,i]=notes_Ob[j,i]*Coef5[i]
moyenne_Ob=np.sum(notes_Ob, axis=1)/4
V5=[]
credit_Ob=[]
for i in range(len(moyenne_Ob)):
    if moyenne_Ob[i] >=10:
        V5.append("V")
        credit_Ob.append(8)

    else:
        V5.append("NV")
        credit_Ob.append(0)
V5
```

# Moyenne Graphique

```python
notes_GR=Tableau[:,13:]
Coef6=[2,2]
for j in range(len(notes_GR)):
    for i in range(2):
        notes_GR[j,i]=notes_GR[j,i]*Coef6[i]
moyenne_GR=np.sum(notes_GR, axis=1)/4
V6=[]
credit_G=[]
for i in range(len(moyenne_GR)):
    if moyenne_GR[i] >=10:
        V6.append("V")
        credit_G.append(8)
    else:
        V6.append("NV")
        credit_G.append(0)
V6
```
